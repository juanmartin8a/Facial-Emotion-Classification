# Data preprocesing

```
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.regularizers import l2
from keras.optimizers import Adam
```

```
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    #zoom_range = 0.2,
    horizontal_flip = True
)
training_set = train_datagen.flow_from_directory(
    'output/training',
    target_size = (48, 48),
    batch_size = 64,
    class_mode="categorical",
)

test_datagen = ImageDataGenerator(
    rescale = 1./255,
)
test_set = test_datagen.flow_from_directory(
    'output/test',
    target_size = (48, 48),
    batch_size = 64,
    class_mode="categorical",
)
```

```
cnn = tf.keras.models.Sequential()

cnn.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu', input_shape=[48, 48, 3], kernel_regularizer=l2(0.01)))
cnn.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn.add(tf.keras.layers.Dropout(0.5))

cnn.add(tf.keras.layers.Conv2D(filters=128, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.Conv2D(filters=128, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn.add(tf.keras.layers.Dropout(0.5))

cnn.add(tf.keras.layers.Conv2D(filters=256, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.Conv2D(filters=256, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.Conv2D(filters=256, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn.add(tf.keras.layers.Dropout(0.5))

cnn.add(tf.keras.layers.Conv2D(filters=512, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.Conv2D(filters=512, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.Conv2D(filters=512, kernel_size=3, activation='relu', padding='same'))
cnn.add(tf.keras.layers.BatchNormalization())
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn.add(tf.keras.layers.Dropout(0.5))

cnn.add(tf.keras.layers.Flatten())

cnn.add(tf.keras.layers.Dense(units=1024, activation='relu'))
cnn.add(tf.keras.layers.Dropout(0.2))
cnn.add(tf.keras.layers.Dense(units=1024, activation='relu'))
cnn.add(tf.keras.layers.Dropout(0.2))
cnn.add(tf.keras.layers.Dense(units=7, activation='softmax'))

cnn.compile(optimizer=Adam(learning_rate=0.001), loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```
print(cnn.summary())
```

```
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
lr_reducer = ReduceLROnPlateau(monitor='val_loss', factor=0.9, patience=3)
#this function monitors the validation loss for signs of a plateau and then
#alter the learning rate by the specified factor if a plateau is detected.
early_stopper = EarlyStopping(monitor='val_acc', min_delta=0, patience=6, mode='auto')
#stops when there is no improvement
checkpointer = ModelCheckpoint('/content/drive/My Drive/Colab Notebooks/Facial Emotion Recognition/Model/weights.hd5', monitor='val_loss', verbose=1, save_best_only=True)
```

```
cnn.fit(x=training_set, validation_data = test_set, epochs = 100, batch_size=64, callbacks=[lr_reducer, checkpointer, early_stopper])
```
